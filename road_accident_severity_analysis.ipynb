```python
import pandas as pd

# Load the Excel file from Google Drive
file_path = '/content/drive/My Drive/Colab Notebooks/dataset1.xlsx'
df = pd.read_excel(file_path)

# Display the first few rows of the dataset
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Convert categorical variables to numerical values
df['Weather'] = label_encoder.fit_transform(df['Weather'])
df['Road Condition'] = label_encoder.fit_transform(df['Road Condition'])
df['Vehicle Type'] = label_encoder.fit_transform(df['Vehicle Type'])

# Display updated dataset
df.head()

```

```python
# Define independent variables and target variable
X = df[['Weather', 'Road Condition', 'Speed Limit (mph)', 'Vehicle Type', 'Time of Day (hour)']]
y = df['Accident Severity']


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Create and train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model using Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')

```

```python
import joblib

# Save the trained model to a file
joblib.dump(model, 'road_accident_severity_model.pkl')

# To reload the model in the future
loaded_model = joblib.load('road_accident_severity_model.pkl')

```

```python
# Example input data for prediction
new_data = pd.DataFrame({
    'Weather': [2],
    'Road Condition': [2],      'Speed Limit (mph)': [110],
    'Vehicle Type': [0],
    'Time of Day (hour)': [19]
})

# Use the loaded model to predict accident severity
severity_prediction = loaded_model.predict(new_data)
print(f'Predicted Accident Severity: {severity_prediction[0]}')
```
